```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the merged dataset
df = pd.read_excel('/Users/kiranmaireddy/Desktop/Semester 4/Result/merged_occupational_and_algorithm_data_with_AF_.xlsx')

# Ensuring 'algorithm_functions' and 'Algorithms' columns are lists (if they are stored as strings)
df['Combined Algorithm Functions'] = df['Combined Algorithm Functions'].apply(eval)
df['Algorithms'] = df['Algorithms'].apply(eval)
```

```python
# Exploding the 'Combined Algorithm Functions' column for easier counting
df_exploded = df.explode('Combined Algorithm Functions')

# Plotting a bar chart showing count of algorithm functions by industry
plt.figure(figsize=(10, 6))
sns.countplot(data=df_exploded, x='Industry', hue='Combined Algorithm Functions', palette="viridis")
plt.title('Count of Algorithm Functions by Industry')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.legend(title='Algorithm Function', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Counting the occurrences of each algorithm function across all occupations
total_function_counts = df_exploded['Combined Algorithm Functions'].value_counts()

# Plotting a pie chart showing distribution of Algorithm Functions across Occupations
plt.figure(figsize=(8, 8))
total_function_counts.plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3"))
plt.title("Distribution of Algorithm Functions Across Occupations")
plt.ylabel("")  
plt.show()
```
